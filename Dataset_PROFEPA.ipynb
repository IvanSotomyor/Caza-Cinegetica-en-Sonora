## Existen columnas en las fechas 2019 en adelante que se agarran como float, se convertirán en object.

```python
print(data_2020.info())
print(data_2021.info())
print(data_2022.info())
print(data_2023.info())
```

```python
data_2019['epocainicio'] = data_2019['epocainicio'].astype(object)
data_2019['epocafin'] = data_2019['epocafin'].astype(object)
```

```python
data_2019.info()
```

```python
data_2019['epocainicio'] = data_2019['epocainicio'].astype(object)
data_2019['epocafin'] = data_2019['epocafin'].astype(object)
data_2020['epocainicio'] = data_2020['epocainicio'].astype(object)
data_2020['epocafin'] = data_2020['epocafin'].astype(object)
data_2021['epocainicio'] = data_2021['epocainicio'].astype(object)
data_2021['epocafin'] = data_2021['epocafin'].astype(object)
data_2022['epocainicio'] = data_2022['epocainicio'].astype(object)
data_2022['epocafin'] = data_2022['epocafin'].astype(object)
data_2023['epocainicio'] = data_2023['epocainicio'].astype(object)
data_2023['epocafin'] = data_2023['epocafin'].astype(object)
```

```python
data_2023.info()
```

```python
merged_data_1 = pd.merge(data_2015, data_2016, how='outer')
merged_data_2 = pd.merge(merged_data_1, data_2017, how='outer')
merged_data_3 = pd.merge(merged_data_2, data_2018, how='outer')
merged_data_4 = pd.merge(merged_data_3, data_2019, how='outer')
merged_data_5 = pd.merge(merged_data_4, data_2020, how='outer')
merged_data_6 = pd.merge(merged_data_5, data_2021, how='outer')
merged_data_7 = pd.merge(merged_data_6, data_2022, how='outer')
merged_data_7.info()
# merged_data_final = pd.merge(merged_data_7, data_2023, how='outer')
# merged_data_final.info()
```

## Quitar comas de los campos CINTIAUTO y CINTIENTREGA para pasarlos de object a float

```python
merged_data_7['CINTIAUTO'] = merged_data_7['CINTIAUTO'].str.replace(',','')
```

```python
merged_data_7['CINTIENTREGA'] = merged_data_7['CINTIENTREGA'].str.replace(',','')
```

```python
merged_data_7['CINTIAUTO'] = merged_data_7['CINTIAUTO'].astype(float)
merged_data_7['CINTIENTREGA'] = merged_data_7['CINTIENTREGA'].astype(float)
```

```python
merged_data_7.info()
```

```python
merged_data_final = pd.merge(merged_data_7, data_2023, how='outer')
merged_data_final.head()
```
